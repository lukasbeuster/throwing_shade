```python
import leafmap
from samgeo import tms_to_geotiff
from samgeo.text_sam import LangSAM
```

```python
image = '../data/clean_data/solar/2315704/2315704_p_56_2022_07_01_rgb.tif'
```

```python
sam = LangSAM()
```

```python
text_prompt = "tree"
```

```python
sam.predict(image, text_prompt, box_threshold=0.24, text_threshold=0.24)
```

```python
sam.show_anns(
    cmap="Greens",
    box_color="red",
    title="Automatic Segmentation of Trees",
    blend=True,
)
```

```python
sam.show_anns(
    cmap="Greens",
    add_boxes=False,
    alpha=0.5,
    title="Automatic Segmentation of Trees",
)
```

```python
sam.show_anns(
    cmap="Greys_r",
    add_boxes=False,
    alpha=1,
    title="Automatic Segmentation of Trees",
    blend=False,
    output="trees.tif",
)
```
